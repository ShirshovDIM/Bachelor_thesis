```python
## comparison for PO and SO whether PO and SO are peforming similarly or SO worser 
hyp_mx = pd.DataFrame({'year': np.arange(2012, 2021)})

def hypothesis(df, pos, alpha = .001, assumption = 'not equal'):
    d = (((df.iloc[pos].PO_std**2 / (df.iloc[pos].PO_counts - 1))
          + (df.iloc[pos].SO_std**2 / (df.iloc[pos].SO_counts - 1)))**.5)
    z_observed = (df.iloc[pos].SO_means-df.iloc[pos].PO_means) / d
    
    z_critical = {.05 : 1.96, .01: 2.56, .001: 3.29}
    
    if assumption == 'not equal':
        if abs(z_observed) < z_critical[alpha]:
            return 'equal'
        else: 
            if df.iloc[pos].SO_means > df.iloc[pos].PO_means:
                return 'better'
            else:
                return 'worser'
```

```python
def data_transformer(SO, PO, year):
    test_so = pd.concat([SO, PO]).loc[:,pd.concat([SO, PO]).columns.map(lambda x: x[-2:]) == str(year)]
    test_so['industry'] = pd.concat([SO, PO]).industry
    test_so['state_owned'] = pd.concat([SO, PO]).state_owned
    check = test_so.loc[:, test_so.count() >= 4000]

    check.columns = ['revenue', 'ebit', 'net_profit', 'assets', 'equity', 'roe_ebit','roce_ebit', 'roa_ebit','roe_net','roce',
                     'roe_net_sm','margin','gross_revenue', 'ebit_margin','return_to_labor','cash_to_labor', 'equity_to_labor',
                     'cap_employed_to_labor','current_coef','liq_coef','equity_liq_coef','payability_coef', 'gearing', 'labor',
                     'industry', 'state_owned']
    # check.columns = ['revenue', 'ebit', 'net_profit', 'assets', 'equity', 'roe_ebit', 'roa_ebit','roe_net',
    #                  'roe_net_sm','margin','gross_revenue', 'ebit_margin','return_to_labor','cash_to_labor',
    #                  'cap_employed_to_labor','current_coef','liq_coef','payability_coef', 'labor', 'industry']

    check.index = range(0, check.shape[0])

    cols = ['roa_ebit', 'gearing','equity', 'labor', 'assets', 'net_profit', 'current_coef', 'cap_employed_to_labor']
    mask = ~check[cols].iloc[:,:-1].apply(lambda x: any(x.isna()), axis = 1)

    check['liability'] = check.gearing * check.equity
    check['roa_ebit'] = check.roa_ebit/100
    check['cer'] = check.cap_employed_to_labor*check.labor.astype(float)/check.assets
    check['gearing'] = check.gearing/100
    check['sfr'] = check.equity/check.assets 
    check.current_coef
    check['os']= check.assets / check.liability / 100

    check = check[mask][(check.gearing != 0) & (~check.cer.isna())]
    check.index = range(0, check.shape[0])
    
    return check
```

```python
def linearity_test(model, y):
    '''
    Function for visually inspecting the assumption of linearity in a linear regression model.
    It plots observed vs. predicted values and residuals vs. predicted values.
    
    Args:
    * model - fitted OLS model from statsmodels
    * y - observed values
    '''
    fitted_vals = model.predict()
    resids = model.resid

    fig, ax = plt.subplots(1,2)
    
    sns.regplot(x=fitted_vals, y=y, lowess=True, ax=ax[0], line_kws={'color': 'red'})
    ax[0].set_title('Observed vs. Predicted Values', fontsize=16)
    ax[0].set(xlabel='Predicted', ylabel='Observed')

    sns.regplot(x=fitted_vals, y=resids, lowess=True, ax=ax[1], line_kws={'color': 'red'})
    ax[1].set_title('Residuals vs. Predicted Values', fontsize=16)
    ax[1].set(xlabel='Predicted', ylabel='Residuals')
    
linearity_test(res_lin_exp, check.roe_ebit) 
```
